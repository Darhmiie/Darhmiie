```python
#monthly sales trend for all branches
pd.date_range(start='2019-01-01', periods=91, freq='D')

# Ensure 'Date' is in datetime format
sales_data['Date'] = pd.to_datetime(sales_data['Date'])

# Extract the month from the Date
sales_data['Month'] = sales_data['Date'].dt.to_period('M')  # Convert to period for monthly grouping

# Group by 'Month' and 'Branch' and sum the sales
monthly_sales = sales_data.groupby(['Month', 'Branch'])['cogs'].sum().unstack().fillna(0)

# Plotting the monthly sales trends for each branch
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o')

# Adding titles and labels
plt.title('Monthly Sales Trend by Supermarket Branch')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Branch')  # Show legend with title
plt.grid()
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
start_date = '2019-01-01'
date_rng = pd.date_range(start=start_date, periods=216 * 4, freq='15T')  # 15-minute intervals

# Define branches
branches = ['Branch A', 'Branch B', 'Branch C']

# Create a DataFrame
sales_data = {
    'Date': np.tile(date_rng, len(branches)),  # Repeat the date range for each branch
    'Branch': np.repeat(branches, len(date_rng)),  # Repeat branch names for each time slot
    'cogs': np.random.randint(50, 300, size=len(branches) * len(date_rng))  # Random sales data
}

sales_data = pd.DataFrame(sales_data)

# Extract the quarter-hour from the Date
sales_data['QuarterHour'] = sales_data['Date'].dt.strftime('%H:%M')  # Format time as HH:MM

# Group by 'QuarterHour' and 'Branch' and sum the sales
quarter_hourly_sales = sales_data.groupby(['QuarterHour', 'Branch'])['cogs'].sum().unstack().fillna(0)

# Plotting the quarterly sales trends for each branch
plt.figure(figsize=(14, 7))
quarter_hourly_sales.plot(kind='line', marker='o')

# Adding titles and labels
plt.title('Quarter-Hourly Sales Trend by Supermarket Branch (216 Hours)')
plt.xlabel('Time of Day (HH:MM)')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate x-ticks for better readability
plt.legend(title='Branch')  # Show legend with title
plt.grid()
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Create a date range for 216 hours starting from a specific date
start_date = '2019-01-01'
date_rng = pd.date_range(start=start_date, periods=216 * 4, freq='15T')  # 15-minute intervals

# Define branches
branches = ['Branch A', 'Branch B', 'Branch C']

# Create a DataFrame
data = {
    'Date': np.tile(date_rng, len(branches)),  # Repeat the date range for each branch
    'Branch': np.repeat(branches, len(date_rng)),  # Repeat branch names for each time slot
    'cogs': np.random.randint(50, 300, size=len(branches) * len(date_rng))  # Random sales data
}

sales_data= pd.DataFrame(data)

# Extract the quarter-hour from the Date
sales_data['QuarterHour'] = sales_data['Date'].dt.strftime('%H:%M')  # Format time as HH:MM

# Group by 'QuarterHour' and sum the sales for all branches
combined_sales = sales_data.groupby('QuarterHour')['cogs'].sum()

# Plotting the combined quarterly sales trend
plt.figure(figsize=(14, 7))
combined_sales.plot(kind='line', marker='o', color='skyblue')

# Adding titles and labels
plt.title('Combined Quarter-Hourly Sales Trend (All Branches)')
plt.xlabel('Time of Day (HH:MM)')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  # Rotate x-ticks for better readability
plt.grid()
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

# PREDICTIVE ANALYSIS

```python
#Use the data to predict sales and customer churn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create a sample DataFrame
np.random.seed(0)
data = {
    'Rating': np.random.randint(1, 11, size=1000),
    'Unit price': np.random.rand(1000) * 100,
    'cogs': np.random.rand(1000) * 500  # Random sales amounts
}

sales_data = pd.DataFrame(data)

# Features and target variable
X = sales_data[['Rating', 'Unit price']]
y = sales_data['cogs']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict sales on the test set
y_pred = model.predict(X_test)

# Calculate and print the mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Create a sample DataFrame with churn information
np.random.seed(0)
data = {
    'Rating': np.random.randint(1, 11, size=1000),
    'Unit price': np.random.rand(1000) * 100,
    'Churn': np.random.choice([0, 1], size=1000)  # 0 = Not churned, 1 = Churned
}

sales_data= pd.DataFrame(data)

# Features and target variable
X = sales_data[['Rating', 'Unit price']]
y = sales_data['Churn']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train a random forest classifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Predict churn on the test set
y_pred = model.predict(X_test)

# Print classification report and confusion matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
